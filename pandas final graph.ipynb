```python
import pandas as pd
df_confirmed = pd.read_csv('./final_df.csv')
df_confirmed.head()
```

```python
df_confirmed.shape
```

```python
# 국가코드 ios2를 통해 국기 이미지를 가져와야 하는데 Namibia의 코드가 NA이다.
# 결측치 : ‘’, ‘#N/A’, ‘#N/A N/A’, ‘#NA’, ‘-1.#IND’, ‘-1.#QNAN’, ‘-NaN’, ‘-nan’, ‘1.#IND’, ‘1.#QNAN’, ‘’, ‘N/A’, ‘NA’, ‘NULL’, ‘NaN’, ‘n/a’, ‘nan’, ‘null’
# keep_default_na=False, na_values='' 사용
country_info = pd.read_csv('./materials/00_materials/COVID-19-master/csse_covid_19_data/UID_ISO_FIPS_LookUp_Table.csv', encoding='utf-8-sig', keep_default_na=False, na_values='')
country_info.head()
```

```python
country_info[country_info['Country_Region'] == 'Namibia']
```

```python
country_info = country_info[['iso2', 'Country_Region']]
country_info.head()
```

```python
# 중복행 제거
country_info.shape # 3560개 행이있다
```

```python
country_info = country_info.drop_duplicates(subset='Country_Region', keep='last')
country_info.shape # 180개로 줄음
```

```python
# 날짜별 국가별 확진자 수와 국가별 iso2 값 병합
doc_final_country = pd.merge(df_confirmed, country_info, how='left', on='Country_Region')
doc_final_country.head()
```

```python
doc_final_country.shape
```

```python
# NaN 값 확인
doc_final_country.isnull().sum()
```

```python
# 특정 조건에 부합하는 행 검색
doc_final_country[doc_final_country['iso2'].isnull()]
```

```python
doc_final_country = doc_final_country.dropna(subset=['iso2'])
doc_final_country[doc_final_country['iso2'].isnull()]
```

```python
#iso2= 'AD'
#flag_link = 'https://www.countryflags.io/' + iso2 + '/shiny/64.png'
#flag_link

def create_flag_link(row):
    flag_link = flag_link = 'https://www.countryflags.io/' + row + '/shiny/64.png'
    return flag_link

doc_final_country['iso2'] = doc_final_country['iso2'].apply(create_flag_link)


doc_final_country.head()
```

```python
# 현재 구조에서는 ios2 정보가 가장 마지막 컬럼이다

cols = doc_final_country.columns.tolist()
cols
```

```python
# 컬럼 위치 변경
cols.remove('iso2')
cols.insert(1, 'iso2') # 1번 인덱스에 추가
```

```python
# 컬럼명 변겅
cols[1] = 'Country_Flag'

doc_final_country.columns = cols
doc_final_country.head()
```

```python
# 파일로 저장
doc_final_country.to_csv('./final_covid_data_for_graph.csv')
```
